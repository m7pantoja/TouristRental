```python
import pandas as pd

df = pd.read_csv('data/clean_spain_2018.csv')
sample_space_size = df.shape[0]

tipo_de_inmueble_spain = df['Tipo_Inmueble'].value_counts()
tipo_de_inmueble_spain_normalized = tipo_de_inmueble_spain/sample_space_size
precio_stats_ccaa = df.groupby(by = 'CCAA')['Precio'].describe()
precio_capacidad_stats_ccaa = df.groupby(by = 'CCAA')['Precio/Capacidad'].describe()
listing_density_ccaa = df['CCAA'].value_counts()
listing_density_ccaa_normalized = listing_density_ccaa/sample_space_size
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

x_tipo_de_inmueble_spain_normalized = np.array(tipo_de_inmueble_spain_normalized.keys())
y_tipo_de_inmueble_spain_normalized = np.array(tipo_de_inmueble_spain_normalized.values)
x_listing_density_ccaa_normalized   = np.array(listing_density_ccaa_normalized.keys())
y_listing_density_ccaa_normalized   = np.array(listing_density_ccaa_normalized.values)
x_precio_mean_ccaa = np.array(precio_stats_ccaa['mean'].keys())
y_precio_mean_ccaa = np.array(precio_stats_ccaa['mean'].values)
x_precio_capacidad_mean_ccaa = np.array(precio_capacidad_stats_ccaa['mean'].keys())
y_precio_capacidad_mean_ccaa = np.array(precio_capacidad_stats_ccaa['mean'].values)

plt.style.use('seaborn-v0_8-deep')

fig1, ax1 = plt.subplots()
ax1.bar(x_tipo_de_inmueble_spain_normalized,y_tipo_de_inmueble_spain_normalized, color = 'brown')
ax1.set_xticklabels(labels = ['Inmueble Completo','Habitación Privada', 'Habitación Compartida'], rotation = 'vertical')
ax1.set_title('Tipos de inmueble ofertados en España para alquiler turístico (%) \n')
fig1.show()

fig2, ax2 = plt.subplots()
ax2.bar(x_listing_density_ccaa_normalized,y_listing_density_ccaa_normalized)
ax2.set_xticklabels(labels = x_listing_density_ccaa_normalized, rotation = 'vertical')
ax2.set_title('Densidad de los anuncios de alquiler turístico en España (%) \n')
fig2.tight_layout()
fig2.show()

fig3, ax3 = plt.subplots(1,2)
ax3[0].bar(x_precio_mean_ccaa,y_precio_mean_ccaa, color = 'darkslategrey')
ax3[0].set_xticklabels(labels = x_precio_mean_ccaa, rotation = 'vertical')
ax3[0].set_title('Precio medio del alquiler turístico \n según comunidades autónomas \n')
ax3[0].set_ylabel('Precio (€)')
ax3[1].bar(x_precio_capacidad_mean_ccaa,y_precio_capacidad_mean_ccaa, color = 'slategrey')
ax3[1].set_xticklabels(labels = x_precio_capacidad_mean_ccaa, rotation = 'vertical')
ax3[1].set_title('Precio medio por persona del alquiler turístico \n según comunidades autónomas \n')
ax3[1].set_ylabel('Precio (€)')
fig3.tight_layout()
fig3.show()
```
